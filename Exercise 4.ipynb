The answer

```python
num = int(input("Please choose a number to divide: "))

listRange = list(range(1,num+1))

divisorList = []

for number in listRange:
    if num % number == 0:
        divisorList.append(number)

print(divisorList)
```

Other answer

```python
num = int(input("Please enter a number for which you want to find all divisors: "))
x = list(range(2,num))

divisorlist = []
for i in x:
    if num%i == 0:
        divisorlist.append(i)
    if len(divisorlist) == 1:
        print("The only divisor for " + str(num) + " is " + str(divisorlist))
    elif len(divisorlist) > 1:
        print("List of divisors for " + str(num) + " are " + str(divisorlist))
    else:
        print("There are no divisors for " + str(num))
```
